```python
df = df.rename(columns={df.columns[0]: 'Datos'})
df
```

```python
# Eliminar las coordenadas y dividir en Comunidad, Provincia y Ciudad
df['Datos'] = df['Datos'].str.replace(r';[^;]+;[^;]+$', '')
df
```

```python
df[['Comunidad', 'Provincia', 'Ciudad']] = df['Datos'].str.split(';', n=2, expand=True)
df
```

```python
df['Provincia'] = df['Provincia'].str.split('/').str[0]
df
```

```python
nuevo_df = df[['Provincia', 'Ciudad']]
nuevo_df['Ciudad'] = nuevo_df['Ciudad'].str.upper()
nuevo_df['Ciudad'] = nuevo_df['Ciudad'].apply(lambda x: unidecode(x))
nuevo_df['Ciudad'] = nuevo_df['Ciudad'].apply(lambda x: re.sub(r'[^\w\s]', '', x))

nuevo_df
```

```python
conteo_ciudades = df['Ciudad'].value_counts()

# Filtrar las ciudades que tengan un conteo mayor o igual a 100
ciudades_filtradas = conteo_ciudades[conteo_ciudades >= 100].index.tolist()

# Filtrar el DataFrame original
df = df[df['Ciudad'].isin(ciudades_filtradas)]
```

```python
df['Ciudad'] = df['Ciudad'].str.replace("L'HOSPITALET DE LLOBREGAT", "HOSPITALET DE LLOBREGAT")
```

```python
df['Ciudad'] = df['Ciudad'].str.upper()
df['Ciudad'] = df['Ciudad'].apply(lambda x: unidecode(x) if isinstance(x, str) else x)
df['Ciudad'] = df['Ciudad'].str.replace('[áäàâ]', 'a', regex=True)  # Reemplazar acentos por "a"
df['Ciudad'] = df['Ciudad'].str.replace('[éêèë]', 'e', regex=True)  # Reemplazar acentos por "e"
df['Ciudad'] = df['Ciudad'].str.replace('[íîìï]', 'i', regex=True)  # Reemplazar acentos por "i"
df['Ciudad'] = df['Ciudad'].str.replace('[óôòö]', 'o', regex=True)  # Reemplazar acentos por "o"
df['Ciudad'] = df['Ciudad'].str.replace('[úûùü]', 'u', regex=True)  # Reemplazar acentos por "u"
df['Ciudad'] = df['Ciudad'].str.replace(r'\([^)]*\)', '', regex=True)  # Eliminar paréntesis y su contenido
df['Ciudad'] = df['Ciudad'].str.replace(r'\scapital$', '', regex=True)

df
```

```python
df['Ciudad'] = df['Ciudad'].str.replace("A CORUNA", "LA CORUÑA")
df['Ciudad'] = df['Ciudad'].str.replace("CASTELLON DE LA PLANA", "CASTELLON")
df['Ciudad'] = df['Ciudad'].str.replace("LAS ROZAS DE MADRID", "LAS ROZAS")
df['Ciudad'] = df['Ciudad'].str.replace("RIVAS VACIAMADRID", "RIVAS")
df['Ciudad'] = df['Ciudad'].str.replace("DONOSTIA", "SAN SEBASTIAN")
```

```python
df['Ciudad'] = df['Ciudad'].str.rstrip(',')
df['Ciudad'] = df['Ciudad'].str.split('-').str[0]
df = df[df['Ciudad'] != '28100']
```
